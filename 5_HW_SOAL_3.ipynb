### Pemahaman Data

```python
# Import Data
import pandas as pd
import numpy as np

# Mempelajari data superstore
superstore_data = pd.read_csv('superstore.csv')
superstore_data
```

```python
# Analisis Deskriptif
superstore_data.describe()
```

### Preprocessing Data


```python
superstore_data.info()

# Tidak ada missing values berdasarkan info
# superstore_data.dropna() untuk menghapus nilai null
```

```python
# Memilih fitur untuk ML
selected_data = superstore_data[['Sales', 'Discount', 'Profit', 'Quantity']]
# X
```

```python
from scipy import stats
z_scores = stats.zscore(selected_data)
print(z_scores)

# superstore_data = superstore_data[(z_scores < 3) & (z_scores > -3)] untuk menangani outlier
```

```python
# One Hot Encoding untuk mengubah data kategori (categorical data) menjadi bentuk yang dapat digunakan oleh algoritma machine learning
# superstore_data = pd.get_dummies(superstore_data, columns=['Category', 'Order ID'], drop_first=True)
```

```python
# Normalisai atau Scalling

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Contoh Min-Max Scaling
scaler = MinMaxScaler()
selected_data = scaler.fit_transform(selected_data)
```

```python
selected_data
```

### Pembagian Data

```python
# Pembagian Dataset
from sklearn.model_selection import train_test_split

X = selected_data.iloc[:,:3]
y = selected_data.iloc[:,3]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=False)
```

### Pembangunan Model dan Evaluasi

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

# Model Prediksi
model = LinearRegression()

# Melatih Model
model.fit(X_train, y_train)

# Membuat Prediksi
y_pred = model.predict(X_test)

# Evaluasi Model Prediksi
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error (MSE): ', mse)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Model Klasifikasi berdasarkan segmen penjualan

# Mencari rata-rata penjualan
average_sales = superstore_data['Sales'].mean()
low_threshold = superstore_data['Sales'].quantile(0.25)

def segmentasi_customer(row):
    if row['Sales'] > average_sales:
        return 'High'
    elif row['Sales'] < low_threshold:
        return 'Low'
    else:
        return 'Medium'

superstore_data['Segment'] = superstore_data.apply(segmentasi_customer, axis=1)

y = superstore_data['Segment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=False)

model = GaussianNB()
model.fit(X_train, y_train)  # X_train adalah fitur-fitur yang terdiri atas penjualan, diskon dan untung, y_train adalah segmentasi
y_pred = model.predict(X_test)

# Mengukur akurasi model
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy}')

# Menampilkan laporan klasifikasi yang mencakup presisi, recall, dan F1-score
print(classification_report(y_test, y_pred))
```
